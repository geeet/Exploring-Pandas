```python
"""Reshaping (Stack and Unstack) - 
Stack : Moves the inner column to the rows of the dataframe. Returns a taller Dataframe
Unstack : Moves the inner rows to the columns of the dataframe. Returns a wider Dataframe. """

mw=df[(df.Edition==2008)&((df.Event=='100m')|(df.Event=='200m'))]
mw
```

```python
g=mw.groupby(['NOC','Gender','Discipline','Event']).size()
g
```

```python
g.plot(kind='barh',color='purple',figsize=(10,6))
```

```python
us=g.unstack(['Discipline','Event'])
us
```

```python
us.stack()
```

```python
us.unstack('NOC')
```

```python
us.unstack('Gender')
```

```python
u=df[df.NOC=='USA']
u
aedm=u.groupby(['Edition','Athlete','Discipline','Medal']).size().unstack('Medal',fill_value=0)
aedm['Total']=aedm['Gold']+aedm['Silver']+aedm['Bronze']
aedm
tu=[group.sort_values('Total',ascending=False)[:1] for year,group in aedm.groupby('Edition')]
tu
top=pd.DataFrame()
for i in tu:
    top=top.append(i)
top
```
